## Feature

Anzahl und Geschlechtertrennung 

```python
temp=[]
amount=[]
male_count=[]
female_count=[]
divers_count=[]
count=0
m=0
f=0
d=0

for a in df.borrower_genders:
     temp=a.split(', ')
     for b in temp:
          count+=1
          if b=='male':
              m+=1
          elif b=='female':
              f+=1
          else:
              d+=1
     amount.append(count)
     male_count.append(m)
     female_count.append(f)
     divers_count.append(d)
     count=0
     m=0
     f=0
     d=0

amount
```

```python
df.insert(len(df.columns),'pers_count', (amount))
df.insert(len(df.columns),'male_count', (male_count))
df.insert(len(df.columns),'female_count', (female_count))
```

```python
df_opti=df.copy()
df_opti.memory_usage(deep=True) 
```

```python
# Wo k√∂nnen floats in ints umgewandelt werden?

#Schleife kontrolliert und speichert die columns von float dtypes
lis=[]
for i in df_opti.columns:
    if df_opti[i].dtype=='float64':
        lis.append(i)
        
#Schleife z√§hlt anzahl der eintr√§ge bei denen keine nachkommastelle auftritt        
#die ersten drei 
a=0
for i in lis:
    for j in df_opti[i]:
        if j%1==0:
            a+=1
            
    print(j, a)
    a=0
```

```python
# in den ersten drei eintr√§gen von lis tritt keine kommazahl auf.
# Wir k√∂nnen diese problemlos umformen.
df_opti[lis[0]]=df_opti[lis[0]].astype(int)
df_opti[lis[1]]=df_opti[lis[1]].astype(int)
df_opti[lis[2]]=df_opti[lis[2]].astype(int)

df_opti.dtypes
```

```python
#Umwandlung von Strings in Categories

for i in df_opti.columns:
    print(i, df_opti[i].nunique())
```

```python

lis=['activity','sector','country_code','country',"region","currency","borrower_genders","continent",'repayment_interval']

for i in lis:
    df_opti.loc[:,i] = df_opti.loc[:,i].astype("category")

df_opti.dtypes
```

```python
reduction = (df.memory_usage(deep=True).sum()-df_opti.memory_usage(deep=True).sum())/df.memory_usage(deep=True).sum()

print(f"{reduction:0f}")
```
